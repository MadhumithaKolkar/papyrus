```python
# Necessary Imports

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books['tagged_description'] = books['isbn10'].astype(str) + ' ' + books['description']
```

```python
print(books["description"])
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python

raw_documents = TextLoader("tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings())
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
import isbnlib

def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)

    books_list = []

    for i in range(0, len(recs)):
        isbn = recs[i].page_content.strip('"').split()[0]

        # Clean the ISBN by removing non-numeric characters and dealing with 'X' at the end
        cleaned_isbn = ''.join(filter(str.isdigit, isbn))

        # Check if it's an ISBN-10 (length 10) and convert to ISBN-13 if necessary
        if len(cleaned_isbn) == 10:
            isbn13 = isbnlib.to_isbn13(cleaned_isbn)
        else:
            isbn13 = cleaned_isbn  # Assume it's already in ISBN-13 format

        books_list.append(isbn13)

    # Return books that match the cleaned ISBN-13s
    return books[books["isbn13"].astype(str).isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach my neighbor science")
```
